```python
# ============================================================================
# SISTEM FUZZY LOGIC UNTUK PREDIKSI KUALITAS UDARA
# ============================================================================

!pip install scikit-fuzzy matplotlib numpy -q

import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
from datetime import datetime

print("="*80)
print(" SISTEM FUZZY LOGIC - PREDIKSI KUALITAS UDARA ".center(80, "="))
print("="*80)
print(f"Waktu Eksekusi: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("="*80)

# ============================================================================
# 1. DEFINISI INPUT DAN OUTPUT VARIABLES
# ============================================================================

print("\n" + "="*80)
print(" DEFINISI MEMBERSHIP FUNCTIONS ".center(80, "="))
print("="*80)

# Input Variable 1: PM2.5 (0-500 µg/m³)
print("\nInput Variable 1: PM2.5 (0-500 µg/m³)")
pm25 = ctrl.Antecedent(np.arange(0, 501, 1), 'pm25')

pm25['baik'] = fuzz.trimf(pm25.universe, [0, 0, 50])
pm25['sedang'] = fuzz.trimf(pm25.universe, [25, 50, 150])
pm25['tidak_sehat'] = fuzz.trimf(pm25.universe, [50, 150, 250])
pm25['berbahaya'] = fuzz.trimf(pm25.universe, [150, 500, 500])

print("  • Baik: trimf [0, 0, 50]")
print("  • Sedang: trimf [25, 50, 150]")
print("  • Tidak Sehat: trimf [50, 150, 250]")
print("  • Berbahaya: trimf [150, 500, 500]")

# Input Variable 2: AQI (0-500)
print("\nInput Variable 2: AQI (0-500)")
aqi = ctrl.Antecedent(np.arange(0, 501, 1), 'aqi')

aqi['baik'] = fuzz.trimf(aqi.universe, [0, 0, 50])
aqi['sedang'] = fuzz.trimf(aqi.universe, [25, 75, 125])
aqi['tidak_sehat'] = fuzz.trimf(aqi.universe, [100, 150, 200])
aqi['berbahaya'] = fuzz.trimf(aqi.universe, [175, 500, 500])

print("  • Baik: trimf [0, 0, 50]")
print("  • Sedang: trimf [25, 75, 125]")
print("  • Tidak Sehat: trimf [100, 150, 200]")
print("  • Berbahaya: trimf [175, 500, 500]")

# Output Variable: Kualitas Udara (0-100)
print("\nOutput Variable: Kualitas Udara (0-100)")
kualitas = ctrl.Consequent(np.arange(0, 101, 1), 'kualitas')

kualitas['sangat_baik'] = fuzz.trimf(kualitas.universe, [0, 0, 25])
kualitas['baik'] = fuzz.trimf(kualitas.universe, [15, 35, 55])
kualitas['sedang'] = fuzz.trimf(kualitas.universe, [45, 60, 75])
kualitas['buruk'] = fuzz.trimf(kualitas.universe, [65, 80, 90])
kualitas['sangat_buruk'] = fuzz.trimf(kualitas.universe, [85, 100, 100])

print("  • Sangat Baik: trimf [0, 0, 25]")
print("  • Baik: trimf [15, 35, 55]")
print("  • Sedang: trimf [45, 60, 75]")
print("  • Buruk: trimf [65, 80, 90]")
print("  • Sangat Buruk: trimf [85, 100, 100]")

# ============================================================================
# 2. VISUALISASI MEMBERSHIP FUNCTIONS
# ============================================================================

print("\n" + "="*80)
print(" VISUALISASI MEMBERSHIP FUNCTIONS ".center(80, "="))
print("="*80)

fig, axes = plt.subplots(nrows=3, figsize=(14, 11))

# Subplot 1: PM2.5
pm25.view(ax=axes[0])
axes[0].set_title('Input Variable 1: PM2.5 Membership Functions',
                  fontsize=13, fontweight='bold', pad=10)
axes[0].set_xlabel('PM2.5 (µg/m³)', fontsize=11, fontweight='bold')
axes[0].set_ylabel('Membership Degree (μ)', fontsize=11, fontweight='bold')
axes[0].grid(True, linestyle='--', alpha=0.5)
axes[0].legend(['Baik', 'Sedang', 'Tidak Sehat', 'Berbahaya'],
               loc='upper right', fontsize=9)

# Subplot 2: AQI
aqi.view(ax=axes[1])
axes[1].set_title('Input Variable 2: AQI Membership Functions',
                  fontsize=13, fontweight='bold', pad=10)
axes[1].set_xlabel('AQI (Air Quality Index)', fontsize=11, fontweight='bold')
axes[1].set_ylabel('Membership Degree (μ)', fontsize=11, fontweight='bold')
axes[1].grid(True, linestyle='--', alpha=0.5)
axes[1].legend(['Baik', 'Sedang', 'Tidak Sehat', 'Berbahaya'],
               loc='upper right', fontsize=9)

# Subplot 3: Kualitas Udara
kualitas.view(ax=axes[2])
axes[2].set_title('Output Variable: Kualitas Udara Membership Functions',
                  fontsize=13, fontweight='bold', pad=10)
axes[2].set_xlabel('Kualitas Udara (Skala 0-100)', fontsize=11, fontweight='bold')
axes[2].set_ylabel('Membership Degree (μ)', fontsize=11, fontweight='bold')
axes[2].grid(True, linestyle='--', alpha=0.5)
axes[2].legend(['Sangat Baik', 'Baik', 'Sedang', 'Buruk', 'Sangat Buruk'],
               loc='upper right', fontsize=9)

plt.tight_layout()
plt.savefig('membership_functions.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Visualisasi membership functions selesai")

# ============================================================================
# 3. DEFINISI FUZZY RULES
# ============================================================================

print("\n" + "="*80)
print(" DEFINISI FUZZY RULES ".center(80, "="))
print("="*80)

print("\nDaftar Rules (Total: 15 rules):\n")

# Rules untuk kombinasi PM2.5 dan AQI
rule1 = ctrl.Rule(pm25['baik'] & aqi['baik'], kualitas['sangat_baik'])
print("  Rule 1 : IF PM2.5 is Baik AND AQI is Baik THEN Kualitas is Sangat Baik")

rule2 = ctrl.Rule(pm25['baik'] & aqi['sedang'], kualitas['baik'])
print("  Rule 2 : IF PM2.5 is Baik AND AQI is Sedang THEN Kualitas is Baik")

rule3 = ctrl.Rule(pm25['baik'] & aqi['tidak_sehat'], kualitas['sedang'])
print("  Rule 3 : IF PM2.5 is Baik AND AQI is Tidak Sehat THEN Kualitas is Sedang")

rule4 = ctrl.Rule(pm25['baik'] & aqi['berbahaya'], kualitas['buruk'])
print("  Rule 4 : IF PM2.5 is Baik AND AQI is Berbahaya THEN Kualitas is Buruk")

rule5 = ctrl.Rule(pm25['sedang'] & aqi['baik'], kualitas['baik'])
print("  Rule 5 : IF PM2.5 is Sedang AND AQI is Baik THEN Kualitas is Baik")

rule6 = ctrl.Rule(pm25['sedang'] & aqi['sedang'], kualitas['sedang'])
print("  Rule 6 : IF PM2.5 is Sedang AND AQI is Sedang THEN Kualitas is Sedang")

rule7 = ctrl.Rule(pm25['sedang'] & aqi['tidak_sehat'], kualitas['buruk'])
print("  Rule 7 : IF PM2.5 is Sedang AND AQI is Tidak Sehat THEN Kualitas is Buruk")

rule8 = ctrl.Rule(pm25['sedang'] & aqi['berbahaya'], kualitas['sangat_buruk'])
print("  Rule 8 : IF PM2.5 is Sedang AND AQI is Berbahaya THEN Kualitas is Sangat Buruk")

rule9 = ctrl.Rule(pm25['tidak_sehat'] & aqi['baik'], kualitas['sedang'])
print("  Rule 9 : IF PM2.5 is Tidak Sehat AND AQI is Baik THEN Kualitas is Sedang")

rule10 = ctrl.Rule(pm25['tidak_sehat'] & aqi['sedang'], kualitas['buruk'])
print("  Rule 10: IF PM2.5 is Tidak Sehat AND AQI is Sedang THEN Kualitas is Buruk")

rule11 = ctrl.Rule(pm25['tidak_sehat'] & aqi['tidak_sehat'], kualitas['sangat_buruk'])
print("  Rule 11: IF PM2.5 is Tidak Sehat AND AQI is Tidak Sehat THEN Kualitas is Sangat Buruk")

rule12 = ctrl.Rule(pm25['tidak_sehat'] & aqi['berbahaya'], kualitas['sangat_buruk'])
print("  Rule 12: IF PM2.5 is Tidak Sehat AND AQI is Berbahaya THEN Kualitas is Sangat Buruk")

rule13 = ctrl.Rule(pm25['berbahaya'] | aqi['berbahaya'], kualitas['sangat_buruk'])
print("  Rule 13: IF PM2.5 is Berbahaya OR AQI is Berbahaya THEN Kualitas is Sangat Buruk")

rule14 = ctrl.Rule(pm25['berbahaya'] & aqi['tidak_sehat'], kualitas['sangat_buruk'])
print("  Rule 14: IF PM2.5 is Berbahaya AND AQI is Tidak Sehat THEN Kualitas is Sangat Buruk")

rule15 = ctrl.Rule(pm25['berbahaya'] & aqi['sedang'], kualitas['buruk'])
print("  Rule 15: IF PM2.5 is Berbahaya AND AQI is Sedang THEN Kualitas is Buruk")

# Kompilasi rules
rules_list = [rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8,
              rule9, rule10, rule11, rule12, rule13, rule14, rule15]

# ============================================================================
# 4. MEMBUAT FUZZY CONTROL SYSTEM (MAMDANI)
# ============================================================================

print("\n" + "="*80)
print(" FUZZY INFERENCE SYSTEM - METODE MAMDANI ".center(80, "="))
print("="*80)

print("\nMembuat Control System dengan Metode Mamdani...")
kualitas_ctrl = ctrl.ControlSystem(rules_list)
kualitas_prediksi = ctrl.ControlSystemSimulation(kualitas_ctrl)
print("✓ Fuzzy Control System berhasil dibuat")

# ============================================================================
# 5. TESTING DENGAN 5 TEST CASES
# ============================================================================

print("\n" + "="*80)
print(" TESTING DENGAN 5 TEST CASES ".center(80, "="))
print("="*80)

# Test cases
test_cases = [
    (10, 20, "Kondisi Sangat Baik - Udara Bersih"),
    (75, 100, "Kondisi Sedang - Aktivitas Normal Aman"),
    (150, 175, "Kondisi Tidak Sehat - Kelompok Sensitif Waspada"),
    (300, 350, "Kondisi Berbahaya - Bahaya Kesehatan Serius"),
    (40, 180, "Kondisi Mixed - PM2.5 Baik tapi AQI Tidak Sehat")
]

results_table = []

for i, (pm25_val, aqi_val, desc) in enumerate(test_cases, 1):
    print(f"\n{'='*80}")
    print(f" TEST CASE {i}: {desc} ".center(80, "="))
    print(f"{'='*80}")

    # Input Values
    print(f"\n📊 INPUT VALUES:")
    print(f"  • PM2.5: {pm25_val} µg/m³")
    print(f"  • AQI  : {aqi_val}")

    # Fuzzy Inference Process
    print(f"\n⚙️  PROSES FUZZY INFERENCE (Mamdani):")

    # Fuzzifikasi
    kualitas_prediksi.input['pm25'] = pm25_val
    kualitas_prediksi.input['aqi'] = aqi_val
    print(f"  [1] Fuzzifikasi: Input dikonversi ke membership degree")

    # Inferensi dan Defuzzifikasi
    print(f"  [2] Inferensi: Evaluasi {len(rules_list)} rules dengan operator AND/OR")
    kualitas_prediksi.compute()
    print(f"  [3] Defuzzifikasi: Metode Centroid untuk hasil crisp")

    # Output
    output = kualitas_prediksi.output['kualitas']

    # Interpretasi
    if output <= 25:
        interpretasi = "SANGAT BAIK"
        emoji = "🟢"
        rekomendasi = "Udara sangat baik. Aman untuk aktivitas outdoor."
    elif output <= 55:
        interpretasi = "BAIK"
        emoji = "🟡"
        rekomendasi = "Kualitas udara dapat diterima. Aktivitas normal."
    elif output <= 75:
        interpretasi = "SEDANG"
        emoji = "🟠"
        rekomendasi = "Kelompok sensitif mulai terpengaruh. Kurangi aktivitas outdoor lama."
    elif output <= 90:
        interpretasi = "BURUK"
        emoji = "🔴"
        rekomendasi = "Tidak sehat untuk semua. Hindari aktivitas outdoor."
    else:
        interpretasi = "SANGAT BURUK"
        emoji = "⚠️🔴"
        rekomendasi = "BAHAYA! Tetap di dalam ruangan. Gunakan masker jika keluar."

    # Display Results
    print(f"\n📈 HASIL DEFUZZIFICATION:")
    print(f"  {'─'*76}")
    print(f"  Nilai Output (Crisp): {output:.4f}")
    print(f"  Kategori Kualitas   : {emoji} {interpretasi}")
    print(f"  {'─'*76}")
    print(f"\n💡 REKOMENDASI:")
    print(f"  {rekomendasi}")

    # Save results
    results_table.append({
        'No': i,
        'PM2.5': pm25_val,
        'AQI': aqi_val,
        'Output': output,
        'Kategori': interpretasi
    })

    # Visualisasi hasil inferensi
    fig, ax = plt.subplots(figsize=(12, 5))
    kualitas.view(sim=kualitas_prediksi, ax=ax)
    ax.set_title(f'Test Case {i}: PM2.5={pm25_val} µg/m³, AQI={aqi_val} → Output={output:.2f} ({interpretasi})',
                 fontsize=12, fontweight='bold', pad=15)
    ax.set_xlabel('Kualitas Udara (Skala 0-100)', fontsize=11, fontweight='bold')
    ax.set_ylabel('Membership Degree', fontsize=11, fontweight='bold')
    ax.grid(True, linestyle='--', alpha=0.5)
    ax.legend(loc='upper right', fontsize=9)
    plt.tight_layout()
    plt.savefig(f'hasil_inferensi_test_case_{i}.png', dpi=300, bbox_inches='tight')
    plt.show()

# ============================================================================
# 6. RINGKASAN HASIL TESTING
# ============================================================================

print("\n" + "="*80)
print(" RINGKASAN HASIL TESTING ".center(80, "="))
print("="*80)

print(f"\n{'No':<4} {'PM2.5':<10} {'AQI':<10} {'Output':<15} {'Kategori':<20}")
print("─"*80)
for result in results_table:
    print(f"{result['No']:<4} {result['PM2.5']:<10} {result['AQI']:<10} "
          f"{result['Output']:<15.4f} {result['Kategori']:<20}")
print("="*80)

# ============================================================================
# 7. BONUS: 3D SURFACE VISUALIZATION
# ============================================================================

print("\n" + "="*80)
print(" BONUS: VISUALISASI 3D SURFACE ".center(80, "="))
print("="*80)

from mpl_toolkits.mplot3d import Axes3D

# Generate grid
pm25_range = np.arange(0, 501, 15)
aqi_range = np.arange(0, 501, 15)
pm25_grid, aqi_grid = np.meshgrid(pm25_range, aqi_range)
kualitas_grid = np.zeros_like(pm25_grid, dtype=float)

print("\nMenghitung nilai kualitas untuk grid 3D...")
for i in range(len(pm25_range)):
    for j in range(len(aqi_range)):
        kualitas_prediksi.input['pm25'] = pm25_grid[j, i]
        kualitas_prediksi.input['aqi'] = aqi_grid[j, i]
        kualitas_prediksi.compute()
        kualitas_grid[j, i] = kualitas_prediksi.output['kualitas']

# Create 3D plot
fig = plt.figure(figsize=(15, 11))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(pm25_grid, aqi_grid, kualitas_grid,
                       cmap='RdYlGn_r', alpha=0.85,
                       edgecolor='gray', linewidth=0.2)

ax.set_xlabel('PM2.5 (µg/m³)', fontsize=12, fontweight='bold', labelpad=10)
ax.set_ylabel('AQI', fontsize=12, fontweight='bold', labelpad=10)
ax.set_zlabel('Kualitas Udara (0-100)', fontsize=12, fontweight='bold', labelpad=10)
ax.set_title('3D Surface Plot: Hubungan PM2.5, AQI, dan Kualitas Udara\n(Fuzzy Inference System - Mamdani)',
             fontsize=14, fontweight='bold', pad=20)

cbar = fig.colorbar(surf, shrink=0.6, aspect=8, pad=0.1)
cbar.set_label('Nilai Kualitas Udara', fontsize=11, fontweight='bold', rotation=270, labelpad=20)

ax.view_init(elev=25, azim=45)

plt.savefig('3d_surface_bonus.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ 3D Surface plot selesai")

# ============================================================================
# PROGRAM SELESAI
# ============================================================================

print("\n" + "="*80)
print(" PROGRAM SELESAI ".center(80, "="))
print("="*80)
print("\n✅ Sistem Fuzzy Logic berhasil diimplementasikan")
print("✅ Metode: Mamdani Fuzzy Inference System")
print("✅ Total Rules: 15 rules")
print("✅ Test Cases: 5 skenario berbeda")
print("✅ Semua visualisasi berhasil dibuat")
print("\n" + "="*80)
```
